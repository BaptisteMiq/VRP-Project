<h2 align="center"><u>Sujet</u></h2> 

<div style="text-align: justify"> Le but de notre étude est de générer une tournée de livraison optimisée sur un réseau routier qui relie de multiples villes. Ce réseau routier est souvent obstrué par de multiples véhicules et ce en fonction des différentes tranches horaires, nous devrons donc faire attention à éviter les voies les plus gênantes lors de notre planification du trajet. De plus notre flotte de camions démarre d'un unique point de départ et chaque camion doit revenir sur ce point dès lors qu'il a fini sa tournée.</div>
<br/>
<div style="text-align: justify"> Notre étude répondra au problème du VRP<sub><em>2</em></sub>. Le "Vehicle Routing Problem" ou "problème de tournées de véhicules" est une classe de problèmes qui a pour objectif de déterminer les tournées d'une flotte de véhicules afin de livrer une liste de clients, de réaliser des tournées d'interventions ou de visites (Exemples: maintenance, réparation, contrôles, visites médicales, commerciales, ...). Le but du problème de VRP est d'optimiser et de minimiser les coûts de trajet d'une flotte de véhicules.</div>

Tout cela en tenant en compte le trafic prévu sur chaque axe pour les différentes tranches horaires. </div>

<b> Recherche opérationnelle approchée : </b> 

Recherche de résolution pour le problème du VRP : 
<div style="text-align: justify">
    

    
Dans le cas de problèmes NP-Complet, il est généralement compliqué de résoudre les instances de grande taille de manière optimale. Nous évitons alors de nous attarder à identifier une solution globale pour nous rabattre pour une solution de « bonne qualité ». Dans l'objectif d'obtenir des solutions dans des temps de calculs raisonnables, on se tourne généralement vers des méthodes approchées à base d'heuristique. Une métaheuristique est un algorithme d’optimisation visant à résoudre des problèmes d’optimisation difficiles pour lesquels on ne connaît pas de méthode classique plus efficace. Une des résolutions les plus adaptées est de commencer par construire une première solution en se basant sur une heuristique que l'on améliorera par la suite grâce à l'aide de multiples heuristiques. Ainsi plusieurs méthodes d'amélioration seront appliquées à chacune des tournées individuellement afin de minimiser le coût global de la solution.
</div> <div style="text-align: justify">
De nombreuses métaheuristiques ont enfin été appliquées à ce problème, comme la recherche tabou, mais aussi les algorithmes génétiques, les algorithmes de colonies de fourmis, les algorithmes de recuit simulé etc. Certains problèmes possédant de très grand nombres de valeurs ont ainsi pu être traités de façon efficace. </div>




<h3 align="center"><u> Etude des Heuristiques et métaheuristiques: </u></h3> 

</div> <div style="text-align: justify"> Comme dit précedement une métaheuristique est un algorithme d’optimisation visant à résoudre des problèmes d’optimisation difficile pour lesquels on ne connaît pas de méthode classique plus efficace. Voici les résultats de nos recherches sur des algorithmes répondant au problèmes VRP: </div>

<em> <u> Recuit Simulé: </u> </em>
</div> <div style="text-align: justify">
Le recuit simulé s'appuie sur l'algorithme de Metropolis-Hastings, qui permet de décrire l'évolution d'un système thermodynamique. Il permet de faire varier une variable de température par apport ou isolation de chaleur. Cette méthode permet de calculer à quelle température un métal type peut se placer dans une configuration atomique solide la plus forte.<br/>
Il permet d’obtenir la meilleure valeur maximum et non pas plusieurs comme l'algorithme de Hill Climbing.
Le recuit est plus optimisé pour faire une myriade de petits calculs car il calcule aléatoirement tous les chemins possibles en sélectionnant le plus optimisé parmi tous les calculs. Hors dans notre cas cela prend plus de ressources car nous partons avec un nombre initial de sommets équivalant à 500 soient une infinité de solutions possibles.
 </div>
 
 <img src="https://upload.wikimedia.org/wikipedia/commons/d/d5/Hill_Climbing_with_Simulated_Annealing.gif" alt="..."  width="500" />

<em> <u> Colonies de fourmis: </u> </em>
</div> <div style="text-align: justify">
L'algorithme de la colonie de fourmis se base sur le comportement de fourmis ou d’autres espèces vivant en colonies, telles que les abeilles ou les termites. Ce sont des insectes sociaux qui ensemble forment un superorganisme. Cet algorithme permet de rechercher des chemins optimaux dans un graphe. Il fonctionne de la manière suivante: Une fourmi éclaireuse parcourt au hasard le graphe, elle identifie les points où passer, puis de multiples fourmis suivent environ le chemin de manière à l’optimiser petit à petit le chemin initial. 
Même si le fait d’améliorer petit à petit notre route est une bonne idée pour de futures améliorations, cet algorithme se base sur l’identification et l’optimisation d’un seul chemin via de multiples passages de fourmis.
</div>

 <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/a/af/Aco_branches.svg/1280px-Aco_branches.svg.png" alt="..."  width="400" />


<em> <u> Tabou: </u> </em>
</div> <div style="text-align: justify">
Métaheuristique itérative qualifiée de recherche locale au sens large souvent utilisée pour résoudre des problèmes complexes et/ou de très grandes tailles.
L'idée de la recherche tabou consiste, à partir d'une position donnée, à en explorer le voisinage et à choisir la position dans ce voisinage qui minimise la fonction “objectif”. En plus de stocker tous les voisins visités pour éviter de revenir sur des solutions déjà explorées (liste tabou). L’algorithme accepte parfois des solutions qui n’améliorent pas toujours la solution courante.
Il permet d'éviter d’être piégé dans un minimum local cependant il exige une gestion de la mémoire de plus en plus lourde en mettant des stratégies de mémorisation complexe. </div>


<em> <u> Hill-Climbing: </u> </em>
</div> <div style="text-align: justify">
La méthode hill-climbing ou méthode d'escalade est une méthode d'optimisation permettant de trouver un optimum local parmi un ensemble de configurations.
Il consiste simplement à partir d’une configuration initiale puis à évaluer les solutions voisines, à chaque fois on choisit la meilleure solution. On recommence l'opération jusqu'à arriver à un optimum local. 
L'algorithme fonctionne de la méthode suivante: 
Premièrement on part d'un élément de notre ensemble de recherche qu'on déclare élément courant. Ensuite on considère le voisinage de l'élément courant et on choisit le meilleur d'entre eux comme nouvel élément courant. Puis on itère jusqu'à convergence ou condition de sortie.
 </div>
 
  <img src="https://static.javatpoint.com/tutorial/ai/images/hill-climbing-algorithm-in-ai.png" alt="..."  width="350" />


<em> <u> Grasp: </u> </em>
</div> <div style="text-align: justify">
Son principe se base sur un mélange entre glouton et aléatoire. Il fait plusieurs départs choisis aléatoirement puis utilise le glouton. Permets de faire un choix parmi tous les minima locaux retenus et donc potentiellement la solution optimale. Déplacement de proche en proche de façon glouton randomisé car le prochain voisin est choisi aléatoirement. C'est une variante de l'algorithme de Hill climbing. </div>


<em> <u> Algorithmes Génétiques: </u> </em>
</div> <div style="text-align: justify">
Les algorithmes génétiques appartiennent à la famille des algorithmes évolutionnistes. Leur but est d'obtenir une solution approchée à un problème d'optimisation, lorsqu'il n'existe pas de méthode exacte (ou que la solution est inconnue) pour le résoudre en un temps raisonnable. Les algorithmes génétiques utilisent la notion de sélection naturelle et l'appliquent à une population de solutions potentielles au problème donné. La solution est approchée par « bonds » successifs, comme dans une procédure de séparation et évaluation (branch et bound), à ceci près que ce sont des formules qui sont recherchées et non plus directement des valeurs.</div>

<em> <u> Exploitation des paramètres de l'instance du problème:  </u> </em>
<div style="text-align: justify">

Afin d'obtenir des valeurs cohérentes nos algorithmes possèdes des variables que l'on peux adapter. Voici les principaux paramètres que l'on fait varier dans nos algorithmes et leurs définitions:
    
<b> Taille de graphe: </b> <br/>
De manière générale, plus la taille du graphe est importante et plus la qualité de la solution est médiocre et plus les complexités spatiales et temporelles sont plus importantes.

<b> Nombre de véhicules:</b> <br/>
Plus le nombre de véhicules est élevé et plus les complexités temporelles et spatiales sont importantes, cependant la qualité des solutions par camions est meilleure. 

<b>Point de départ / dépôt: </b> <br/>
Le point de départ peut avoir un impact sur la qualité de la solution trouvée, selon le graphe généré.
    
<br/>
<b> Impact des paramètres sur l’algorithme du tabou: </b>
<ul>
    <li> Taille de la liste tabou: Plus la liste tabou est importante, plus l’algorithme est en capacité de tester de nouvelles solutions et par conséquent de trouver éventuellement de meilleurs résultats. Une liste tabou plus importante aura également pour effet de provoquer une complexité spatiale plus grande. </li>
    <li> Nombre d’itérations: Représente le nombre de fois que l’on va exécuter l’algorithme tabou. Plus ce nombre d’itérations est important, plus l’algorithme aura de chances de trouver de meilleurs chemins.</li>
</ul>
    
<br/>
<b> Impact des paramètres sur l’algorithme de colonies de fourmies: </b>
<ul>
    <li> Visibilité = Équivalent de la pondération, chaque fourmis déterminera son chemin via la visibilité, les phéromones et un chance de déviance.  </li>
    <li> Phéromones = Marque laissée par le passage des fourmis, ils sont plus présents sur un chemin avec meilleur rendement. Les fourmis passent plus sur un chemin qui a une bonne pondération phéromonale. </li>
    <li> Nombre de fourmis = Définit la taille de la colonie de fourmis. Plus la colonies de fourmis est importante et plus l’identification d’un chemin optimal sera rapide. </li>
    <li>Évaporation = Contrôle la vitesse à laquelle les phéromones disparaissent, plus elle est élevée et plus les phéromones mettront de temps à disparaître. De grande valeurs ralentissent l’identification d’un chemin optimal car les fourmis auront plus de mal à remplacer le chemin avec le plus phéromones par un nouveau chemin plus optimisé.   </li>
</ul>

<br/>
<b> Impact des paramètres sur l’algorithme génétique : </b>
<ul>
    <li> La population : la population désigne le nombre d'individus de chaque génération, une plus grande population consiste en une plus grande diversité d'individus et donc plus d'individus auxquels il est nécessaire d'effectuer des mutations. Augmenter le nombre d'individus par population induit un plus grand temps d’exécution mais augmente significativement la diversité génétique de la population (afin de quitter les optima locaux). </li>
    <li> Le nombre de génération à effectuer : ce paramètre désigne le nombre de fois où l’on effectue des modifications sur génome des individus de la population. Augmenter le nombre de générations permet de converger vers un meilleur résultat au détriment du temps de calcul qui nécessite plus d’itération.</li>
    <li> Le taux de mutation : il détermine le nombre de fois où l’on doit effectuer une mutation (permutation de sommet dans la liste de trajet). Augmenter ce paramètre permet d'agrandir de domaine de recherche des solutions en plus d’augmenter indirectement la diversité du génome de la population, mais fait converger bien lentement vers les solutions optimales. </li>
</ul>
    
<br/>
<b> Impact des paramètres sur l’algorithme recuit: </b>
<ul>
<li> Nombre d'itération: Valeur d'itération initial pour générer une combinaison semi-optimiser d’entrée pour la fonction du recuit simulé. </li>
<li> La température: Valeur permettant de trouver un chemin de plus en plus optimisé se rapprochant d’un optimum au minimum local en faisant décroitre cette valeur. </li>
<li> Dépôt: Paramètre d’entrée pour choisir un dépôt initial pour l’ensemble des camions et ainsi générer la meilleur route pour chaque camion à partir de ce point source. </li>
<li> Itération d’optimisation: Paramètre permettant d'affiner le résultat obtenu en réitérant le recuit simulé sur un chemin déjà optimisé. </li>
<li> Graph: Paramètre d’entrée permettant de trouver le chemin le plus optimisé à partir d’une matrice de pondération.
 </li>
</ul>

<em> <u> Qualité résultats Tabou et Génétique:  </u> </em>



<div style="text-align: justify">
<b> Analyse des résultats: </b> <br/>
On remarque que pour de petites solutions (inférieure à 30 sommets) l’algorithme génétique génère de meilleures solutions. 
Cela s'explique par l'utilisation d’un nombre de graphe changeant mais par un nombre de générations et de population qui reste fixe. Pour améliorer la solution, il faudrait augmenter ces valeurs mais cela prendrait trop de temps à s’exécuter. Un autre problème de l’algorithme génétique est qu’au début de son exécution il se base sur une solution initiale aléatoire. </div>


<em> <u> Mélange de Métaheuristique pour identifier un meilleur résultat (en qualité des résultats): </u> </em>
<div style="text-align: justify">
Pour pallier aux problèmes de l'algorithme génétique nous avons choisi d'utiliser la solution initiale générée par l’algo du tabou. Nous avons eu l'idée de créer cette nouvelle version car les complexités temporelle et spatiale du tabou sont quasiment nulles pour les instances de graphes observées. Ce nouvel algorithme se nomme "Tabou-Génétique" ou "Taboulé OGM". 
<br/> </div>


<em> <u> Description du nouvel algorithme "Tabou-Génétique": </u> </em>
<div style="text-align: justify">
L’algorithme Tabou-Génétique comme son nom l’indique se base sur les métaheuristiques de la recherche tabou et les algorithmes génétiques. Il s'agit de la succession de l’exécution des deux algorithmes, le résultat de la recherche tabou permettant d’initialiser la population de l’algorithme génétique. Il vient alors améliorer les solutions retenues par la recherche tabou en utilisant la sélection naturelle, permettant de trouver un résultat aussi bien voire plus que la recherche tabou à elle seule. <h3 align="center"><u>Modèle statistique</u></h3>

<div style="text-align: justify">
Une fois les données acquises, nous pouvons appliquer un modèle statistique linéaire et créer une droite de régression pour chaque arête. Cette droite est issue d'une fonction affine qui détermine notre modèle statistique prédictif. A partir de cette fonction nous pouvons alors prédire une estimation du trafic sur un axe routier (ou arête) en fonction du temps.

Exemple sur les premieres arêtes :

```python
from statsmodels.graphics import tsaplots
import statsmodels.api as sm

def generateRegressionArrete(numero, creneaux, timePonderation, printGraph):
    if str(creneaux) == "m":
        minutesMax = 540
        minutesMin = 420
    elif str(creneaux) == "s":
        minutesMax = 1140
        minutesMin = 1020
    elif str(creneaux) == "j":
        minutesMax = 1440
        minutesMin = 0
        
    arreteData = collection_trafic_stamped.find({"num_arete": numero})
    arreteList = []
    x = []
    y = []
    moyenne = 0
    for i in arreteData:
        time = str(datetime.time(i['date']))
        t=time.split(':')
        total_minutes= int(t[0])*60+int(t[1])*1 +int(t[2])/60
        if minutesMin < total_minutes < minutesMax:
            moyenne += i['nb_vehicules']
            x.append(total_minutes)
            y.append(i['nb_vehicules'])
    #Récupération du coeficient de droite de regression
    coef = np.polyfit(x, y, 1)
    poly1d_fn = np.poly1d(coef)
            
    #Corélation matrix
    correlation_matrix = np.corrcoef(x, y)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    if(printGraph): 
        print('r_carre: ', r_squared)
        print("Regression linéaire de l'arête", numero)
        graphRegression(x, y, poly1d_fn, 420, 540)
        print('Graphique résiduel du trafic de l\'arête', numero)
        graphResidu(x, y)
    
    #Récupération du a + b pour calculer notre droit ax+b
    coefSplitA, coefSplitB = splitCoef(poly1d_fn)
    
    return  Y(timePonderation, coefSplitA, coefSplitB), r_squared
    

def graphRegression(x, y, poly, sizeMin, sizeMax):
    #Récupération du a + b pour calculer notre droit ax+b
    coefSplitA, coefSplitB = splitCoef(poly)
    #Plot de nos points + dorite de régression  
    plt.figure(figsize=(15,6))
    plt.plot(x, y, "bo", label="Regression Line")
    plt.plot(x, [Y(X, coefSplitA, coefSplitB ) for X in x], '-r', label="Coeficient de droite directrice")
    plt.xlim(sizeMin, sizeMax)
    plt.legend()
    plt.show()
    
def graphResidu(x, y):
    plt.figure(figsize=(15,6))
    X = np.append(arr = np.ones((len(x), 1)).astype(int), values = np.array([x]).T, axis = 1)
    # Entrainement du modèle
    regressor_OLS = sm.OLS(endog = y, exog = X).fit()
    
    # Affichage des résidus
    fig, ax = plt.subplots()
    ax.scatter(X[:,1], #Les résidus
    regressor_OLS.resid, alpha=0.3)
    plt.show()
    
#Retourne le a ax+b d'un point x
def Y(x, a, b):
    return (a*x + b)

#Split du coef directeur en string pour récupérer le a et le b de ax+b en float
def splitCoef(coef):
    a = str(coef).split('x')
    b = a[1].split(' ')
    return float(a[0]), float(b[1]+b[2])

ponderation = []
print("Pour l'arête 0:")
generateRegressionArrete(0, "m", 450, True)
print("Pour l'arête 5:")
generateRegressionArrete(5, "m", 450, True)

rcarreTotal = 0
for i in range(500):
    ponderationVal,rcarre = generateRegressionArrete(i, "m", 450, False)
    ponderation.append(ponderationVal)
    rcarreTotal += rcarre 
rcarreMoy = rcarreTotal/500
print("Moyenne des r carrés:" + str(rcarreMoy))
```

<div style="text-align: justify">
On peut voir avec sur les exemples ci-dessus que les modèle statistique et la regression linéaire sont cohérents, de plus nous avons calculé le coefficient de détermination afin de vérifier la pertinence du modèle linéaire.</div>
<div style="text-align: justify">
Pour ce qui est des coefficient de détermination (ou R²), on obtient dans l'exemple des valeurs supérieures à 0.5 ce qui signifie que le modèle est cohérent. Nous avons enfin généralisé le modèle sur toutes les arêtes et fait une moyenne globale de tous le coefficient de détermination. Cette fois, le résultat montre la pertinence du modèle sur tout notre jeu de données.

<h3 align="center"><u>Genération des graphes</u></h3>

<div style="text-align: justify">
Maintenant que nous avons généré un modèle statistique par arête, nous pouvons créer notre graphe utilisant les données réelles de prédiction.</div>
<div style="text-align: justify">
Nous pouvons donc à l'aide du temps (le prédicteur) déterminer la valeur prédictive: la pondération d'arête.

```python
import pandas as pd
def dataGeneratorReel(proba, nbSommet, areteList, doPrint):
    b = np.random.choice((1, 0), size=(nbSommet,nbSommet), p=[proba, 1-proba])
    mat_sym = np.logical_or(b, b.T)
    #seconde instance de tableau
    newTab = [0] * nbSommet
    
    #Out of aretes list range exception
    if len(areteList) <= (nbSommet*(nbSommet - 1))/2:
        print("Error: Nombre de sommet trop important pour nombre d'aretes disponible")
        return
    
    #generation du tableau à partir du nombre de sommet
    for i in range(nbSommet):
            newTab[i] = [0] * nbSommet
            for j in range(nbSommet):
                newTab[i][j] = 0
    #Generation des nombre de prevision sur les aretes
    indexAreteList = 0
    for indexX,x in enumerate(mat_sym.astype(int)):
        for indexY,y in enumerate(x):
            if newTab[indexX][indexY] == 0 and y == 1 and indexY != indexX:
                areteWeight = round(areteList[indexAreteList], 3)
                #génère le random de prevision dans les deux points du tableaux
                newTab[indexX][indexY] = areteWeight
                newTab[indexY][indexX] = areteWeight
                indexAreteList += 1
        
    if doPrint:
        pprint.pprint(newTab)
    return newTab
    
    
graph = dataGeneratorReel(1, 32, ponderation, False)
a = pd.DataFrame(graph)
a
```
