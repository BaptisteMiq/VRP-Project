# Generate algorithm stats

## Imports and retrieve datasets

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from stats.stats import GetStats, showStats
from algorithms.tabu import tabu
from tools.graphGeneration import randomDataGenerator, csvToGraph

g20 = csvToGraph("dataset/ran-20x20.csv")
g40 = csvToGraph("dataset/ran-40x40.csv")
g80 = csvToGraph("dataset/ran-80x80.csv")
g140 = csvToGraph("dataset/ran-140x140.csv")
g500 = csvToGraph("dataset/ran-500x500.csv")

# Default graph sizes for generated stats
minGraphSize = 4
maxGraphSize = 20
paddingGraph = 4
maxGraph = g20 # Must be > to maxGraphSize

print("This config will generate", len(range(minGraphSize, maxGraphSize, paddingGraph)), "graphs per algorithm.")

print("Using max graph:")
p = pd.DataFrame(maxGraph)
p
```

## Stats of 4 main algorithms

```python
x, dataset = GetStats(minGraphSize, maxGraphSize, paddingGraph, ["tabou", "genetic", "fourmi", "recuit"], "", False, maxGraph)
showStats(x, dataset, maxGraphSize, "worst")
showStats(x, dataset, maxGraphSize, "ram")
showStats(x, dataset, maxGraphSize, "time")
```

## Closer look on the 2 best algorithms (Tabu & Genetic)

```python
x, dataset = GetStats(minGraphSize, maxGraphSize, paddingGraph, ["tabou", "genetic"], "", False, maxGraph)
showStats(x, dataset, maxGraphSize, "worst")
```

## Mixing Tabou and Genetic

```python
x, dataset = GetStats(minGraphSize, maxGraphSize, paddingGraph, ["tabou", "genetic", "tabouOGM"], "", False, maxGraph)
showStats(x, dataset, maxGraphSize, "worst")
showStats(x, dataset, maxGraphSize, "ram")
showStats(x, dataset, maxGraphSize, "time")
```

## Comparison with the Google Tabou algorithm

```python
gx1000 = np.multiply(g500, 1000).astype(int).tolist() # The OR-Tools only accept distances and time matrixes, not normalized values 
x, dataset = GetStats(minGraphSize, maxGraphSize, paddingGraph, ["tabou", "genetic", "tabouOGM", "google"], "AUTOMATIC", False, gx1000)
showStats(x, dataset, maxGraphSize, "worst")
showStats(x, dataset, maxGraphSize, "ram")
showStats(x, dataset, maxGraphSize, "time")
```
