# Generate algorithm stats

## Imports and retrieve datasets

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from stats.stats import GetStats, showStats
from algorithms.tabu import tabu
from tools.graphGeneration import randomDataGenerator, csvToGraph

g20 = csvToGraph("dataset/ran-20x20.csv")
g40 = csvToGraph("dataset/ran-40x40.csv")
g80 = csvToGraph("dataset/ran-80x80.csv")
g140 = csvToGraph("dataset/ran-140x140.csv")
g500 = csvToGraph("dataset/ran-500x500.csv")

# Tailles des graphes utilisés pour générer les statistiques
minGraphSize = 4
maxGraphSize = 140
paddingGraph = 4
maxGraph = g140

print("Cette configuration génèrera", len(range(minGraphSize, maxGraphSize, paddingGraph)), "graphes par algorithme.")

print("Using max graph:")
p = pd.DataFrame(maxGraph)
p
```

```python
x, dataset = GetStats(minGraphSize, maxGraphSize, paddingGraph, ["tabou", "lower"], "", False, maxGraph)
showStats(x, dataset, maxGraphSize, "worst")
```
